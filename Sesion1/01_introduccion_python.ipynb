# Listas 

```python
# Lista de Enteros
my_list_1 = [3, 5, 2]
print(my_list_1)

# Lista vacia
my_list_2 = []
print(my_list_2)

# Lista Mixta
my_list_3 = [3, 'Hello', 44.44]
print(my_list_3)

# Lista Anidada
my_list_4 = ['apple', [3, ['a', 2, 3],8], ['x']]
print(my_list_4)
```

```python
my_list_4.insert(1, 30)
print(my_list_4)
```

```python
my_list = ['x', 'l', 'm', 'a']

#primer elemento
print(my_list[0])

#tercer elemento
print(my_list[2])

#lista anidad
n_list = ['Happy', [6, 2, 1, 8]]

print(n_list[0][1])
print(n_list[1][1])
```

### Tupla

```python
# tuple empty
my_tuple=()
print(my_tuple)

# tuple de enteros
my_tuple = (3, 2, 5)
print(my_tuple)

# mixta
my_tuple = (3, 'Hola Mundo', 5.9)
print(my_tuple)

# mixta
my_tuple = (3, 'Hola Mundo', (6, 5, 6), 5.9)
print(my_tuple)
```

```python
# la tupla no es mutable
my_tuple[0]= 'banana'
```

```python
my_tuple = ('p', 'e', 'r', 'm', 'i', 't')

print(my_tuple[4])
print(my_tuple[1])
```

```python
my_tuple = ('p', 'e', 'r', 'm', (5,4,3), 't')

print(my_tuple[4][2])
```

## Diccionario

```python
# Diccionario Vacio
my_dict = {}

my_dict = {
    1: 'apple',
    2: 'pera'
}

print(my_dict)

my_dict_2 = dict({1:'apple', 2:'pera'})
print(my_dict_2)
```

```python
my_dict_2 = {
    'name': 'Jhon',
    1: {'edad': 12}
}

my_dict_2
```

```python
my_dict = {
    'name': 'Jack',
    'age': 26
}

# Obtener un ejemento
print(my_dict['name'])

# metodo2 :
print(my_dict.get('age', 0))
```
