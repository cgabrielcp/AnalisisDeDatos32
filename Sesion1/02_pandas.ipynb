```python
# Importamos libreria
import pandas as pd

# Tipos de datos
#   lista: []
#   diccionario: {}
#   tupla: ()
```

```python
data = [['Juan', 25], ['Pedro', 34], ['Luis', 45]]

df = pd.DataFrame(data, columns = ['Nombre', 'Edad'])
```

```python
df
```

```python
# Crear un data frame a partir de un diccionario

data = {
    'Nombre': ['Juan', 'pedro', 'Luis'],
    'Edad': [25, 26, 30]
}

df = pd.DataFrame(data)
```

```python
df
```
